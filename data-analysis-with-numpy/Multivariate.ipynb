```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ipl=pd.read_csv("/content/class_ipl_22-26_dataset.csv")
ipl.head()
```

```python
ipl.shape
ipl.columns
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='team1_wickets',
    y='team1_runs',
    data=ipl.tail(74)
)

plt.title('Team1 Wickets vs Team1 Runs')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
     x='team1',
    y='team1_runs',
     hue='winner',
     data=ipl.tail(74)
).tick_params(axis='x',rotation=90)

plt.title('Win by Wickets vs Win By Runs by winner')
plt.show()
```

```python
multi_var_cols = ['team1_runs', 'team2_runs', 'win_by_runs', 'win_by_wickets', 'season', 'result_type']
multi_var_df = ipl[multi_var_cols].head(100)

sns.pairplot(multi_var_df, hue='result_type', diag_kind='kde') # Changed hue and diagonal plots
plt.suptitle('Multivariate Analysis: Runs, Wickets, Season by Result Type', y=1.02)
plt.show()
```

```python
numerical_cols = ['season', 'win_by_runs', 'win_by_wickets', 'team1_runs']
corr_df = ipl[numerical_cols]

corr_matrix = corr_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(
corr_matrix,
annot=True,
cmap='coolwarm',
fmt=".2f"
)
plt.title('Correlation Heatmap of Selected IPL Match Data Variables')
plt.show()
```
