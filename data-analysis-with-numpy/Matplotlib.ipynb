Matplotlib Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt


df =pd.read_csv('/content/class_ipl_22-26_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df['team1_wickets'].max()
```

```python
print(df['team2_wickets'].idxmax())
```

```python
plt.plot(df['team1_wickets'].tail(74))
plt.xlabel('Match Index')
plt.ylabel('Team 1 Runs')
plt.title('Team 1 runs trend (Last 74 Matches)')
plt.show()
```

```python
plt.plot(df['team2_runs'].tail(74),'g--',label='Team 2 Runs')
plt.xlabel('Match Index')
plt.ylabel('Team 2 Runs')
plt.title('Team 2 Runs Trend (Last 74 Matches)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['team2_runs'].tail(74),label='Team runs ',color='purple')
plt.plot(df['team1_wickets'].tail(74),label='Team wickets',color='orange')
plt.xlabel('Runs')
plt.ylabel('wickets')
plt.title('Team Runs Vs Team wickets')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['team1_runs'].tail(74),label='Team1 Runs',color='green')
plt.plot(df['team2_runs'].tail(74),label='Team2 Runs',color='maroon')
plt.xlabel('Match Index')
plt.title("Team1 Runs Vs Team2 Runs Trend (Last 74 Matches)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
match_indices=df.index[-74:]


plt.scatter(match_indices, df['team1_runs'].tail(74),c='blue',label='Team 1 Runs')
plt.scatter(match_indices, df['team2_runs'].tail(74),c='red',label='Team 2 Runs')
plt.xlabel('Match Index')
plt.ylabel('Runs')
plt.title('Team 1 Runs and Runs per Match (Last 74 Matches)')
plt.legend()
plt.show()
```

```python
plt.hist(df['team2_wickets'].tail(74),bins=10)
plt.xlabel('Wickets')
plt.ylabel('Number of Matches')
plt.title('Distribution of Team 2 Wickets')
plt.show()
```

```python
toss_decisions=df['toss_decision'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(toss_decisions.index,toss_decisions.values,color=['skyblue','lightcoral'])
plt.xlabel('Toss Decision')
plt.ylabel('Number of Matches')
plt.title('Distribution of Toss Decisions')
plt.show()
```

```python
team_wins=df['winner'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(team_wins.index,team_wins.values,color=['skyblue','lightcoral'])
plt.xlabel('Winning Teams')
plt.ylabel('Number of Matches')
plt.title('Distribution of winning Teams')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.boxplot(df['team2_runs'])
plt.ylabel('Team 2 Runs')
plt.title('Distribution of Team 2 Runs')
plt.show()
```

```python
df['team2_runs'].describe()
```

```python
result_type_counts =df['winner'].value_counts()


plt.figure(figsize=(8,8))
plt.pie(result_type_counts.values,labels=result_type_counts.index,pctdistance=1,radius=2)
plt.title('Distribustion of Match Result Types')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

wicket_data = {
    '0-3': len(df[(df['team1_wickets'] <= 3)]),
    '4-7': len(df[(df['team1_wickets'] >= 4) & (df['team1_wickets'] <= 7)]),
    '8+': len(df[(df['team1_wickets'] >= 8)])
}

plt.pie(
    wicket_data.values(),
    labels=wicket_data.keys(),
    autopct='%1.1f%%'
)

plt.title("Team 1 wicket Distribution")
plt.show()
```
