```python
virat_runs =np.array([45,72,33,61,88,54,40,76,92,58])
rohit_runs=np.array([55,70,30,80,93,36,45,90,68,42])

rohit_total=np.sum(rohit_runs)
kohli_total=np.sum(virat_runs)

print("virat Total Runs:",kohli_total)
print("Rohit Total Runs:",rohit_total)

if kohli_total > rohit_total:
    print("Virat Kohli scored more total runs")

elif rohit_total > kohli_total:
    print("Rohit scored more total runs")
else:
    print("Both scored equal total runs")

kohli_avg=np.mean(kohli_runs)
rohit_avg=np.mean(rohit_runs)

avg_diff =abs(kohli_avg-rohit_avg)

print("Kohli Average:", kohli_avg)
print("Rohit Average:",rohit_avg)

print("Difference in Average Runs:",avg_diff)

kohli_better_matches =np.sum(kohli_runs>rohit_runs)

print("Matches where kohli scored more than Rohit:",kohli_better_matches)
```

```python
team_runs=np.array([[45,60,30],[50,70,40],[20,30,50],[60,40,80],[30,20,40]])

batter_totals =np.sum(team_runs,axis=1)
print("Total Runs of Each Batter:")
print(batter_totals)

match_totals =np.sum(team_runs,axis=0)
print("Match wise Team Totals:")
print(match_totals)

best_batter =np.argmax(batter_totals)+1

print("Beast Performing Batter: Batter",best_batter)
print("Runs:",batter_totals[best_batter-1])
```

```python
scores =np.array([45,72,33,61,88,54,40,76,92,58])

mean_scores =np.mean(scores)


variance =np.var(scores)

std_dev=np.std(scores)

print("Mean:",mean_scores)
print("Variance:",variance)
print("standard Deviation:",std_dev)

if std_dev <20:
    print("Player is Consistent")
else:

    print("Player is Inconsistent")
```

```python
runs =np.array([1,4,0,6,2,1,4,0,1,6,2,1])

total_runs =np.sum(runs)

boundary_runs = np.sum(runs[(runs == 4) | (runs == 6)])

boundary_percentage=(boundary_runs / total_runs)*100

print("Total Runs:",total_runs)
print("Boundary Runs:",boundary_runs)
print("Percentage of Runs through Boundaries:",round(boundary_percentage,2))


print()
```
